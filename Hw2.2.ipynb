```python
# Використовуючи метод groupby, виконайте групування за стовпчиком "Посада
```

```python
gr_by_posada = python_data.groupby(['Должность'])
gr_by_posada.count().index
gr_by_posada.mean(numeric_only=True).sort_values("salary", ascending=False)
```

```python
# Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада",
# виконайте агрегацію даних за допомогою методу agg
# і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
```

```python
gr_by_posada['salary'] \
  .agg(['min', 'mean', 'max'])\
  .sort_values("mean", ascending=False) \
  .apply(lambda x: pd.to_numeric(x, errors='coerce').astype(int) )
```

```python
spec_stat = gr_by_posada['salary'].agg(['min', 'max'])
spec_stat.head()
```

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць.
# Використовуйте її для методу apply та створіть новий стовпчик "avg"
```

```python
gr_by_posada['salary'].mean(numeric_only=True)
```

```python
gr_by_posada['salary'].agg('mean')
```

```python
spec_stat = gr_by_posada['salary'].agg(['min', 'max'])
print(spec_stat)
gr_by_posada['salary'].describe()
```

```python
# USE REFERENCE

def fill_avg_salary(s, ref_df):
  avg = None
  try:
    sp = s.get('Должность')
    if sp:
      stat_sp = ref_df.loc[sp]
      avg=int((stat_sp["min"] + stat_sp["max"])/2)
  finally:
    s['avg']=avg
    return s


a=df.apply(fill_avg_salary, args=[spec_stat], axis=1)
a[a.columns[[0,1,2,19,20]]].head(2)
```

```python
# USE GLOBAL ENV

def fill_avg_salary(s):
  avg = None
  try:
    sp = s.get('Должность')
    if sp:
      stat_sp = spec_stat.loc[sp]
      avg=int((stat_sp["min"] + stat_sp["max"])/2)
  finally:
    s['avg']=avg
    return s


a=df.apply(fill_avg_salary, axis=1)
a[a.columns[[0,1,2,19,20]]].head(2)
```

```python
def fill_avg(s):
  avg = None
  try:
      avg=int((s["min"] + s["max"])/2)
  finally:
    s['avg']=avg
    return s

b=spec_stat.apply(fill_avg, axis=1)
b
```
